```
import pandas as pd
#df_AGG = pd.read_csv('ad_matched_queries_AGG.csv')
#df_AGG

df_BYDATE = pd.read_csv('ad_matched_queries_BYDATE.csv')
df_BYDATE.head(10)
```

```
## keywords dataset
#df_keywords = pd.read_csv('Search keyword report 20190916.csv')
#df_keywords
```

```
df_BYDATE['transaction/clicks'] = df_BYDATE['transactions'] / df_BYDATE['clicks']
```

```
df_BYDATE['created_accounts'].value_counts()
```

```
import numpy as np
df_BYDATE['create_account1'] = np.where(df_BYDATE["created_accounts"] > 0, 1, 0)
df_BYDATE['create_account1'].value_counts()
```

```
from sklearn.model_selection import train_test_split
query_train, query_test, created_train, created_test = train_test_split(df_BYDATE['ad_matched_query'], df_BYDATE['create_account1'] ,test_size=0.2, random_state=42)
```

```
from keras.preprocessing.text import Tokenizer

t = Tokenizer(num_words=20000)
t.fit_on_texts(query_train)

print(t.word_counts)
print(t.document_count)
print(t.word_index)
print(t.word_docs)
```

```
t.fit_on_texts(query_test)

print(t.word_counts)
print(t.document_count)
print(t.word_index)
print(t.word_docs)
```

```
sequences_train = t.texts_to_sequences(query_train)
sequences_test = t.texts_to_sequences(query_test)
sequences_train
```

```
#encoded_queries = t.texts_to_matrix(df_BYDATE['ad_matched_query'], mode='binary')
#print(encoded_queries)
```

```
from keras_preprocessing import sequence

MAX_SEQUENCE_LENGTH = 500

max_length = len(max(sequences_train, key=len))
if max_length > MAX_SEQUENCE_LENGTH:
  max_length = MAX_SEQUENCE_LENGTH

#make sequennces same length
sequences_train = sequence.pad_sequences(sequences_train, maxlen=max_length)
sequences_test = sequence.pad_sequences(sequences_test, maxlen=max_length)

#sequences_train
#sequences_test
t.word_index
```

```
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(sequences_train,created_train)
created_pred = nb.predict(sequences_test)

from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(created_test, created_pred, average='weighted')
```

```
#confusion matrix
from sklearn import metrics

metrics.confusion_matrix(created_test, created_pred)
```

```
#logistic regression

from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(sequences_train,created_train)
created_pred = logreg.predict(sequences_test)
```

```
created_pred_prob = logreg.predict_proba(sequences_test)[:, 1]
created_pred_prob
```

```
precision_recall_fscore_support(created_test, created_pred, average='weighted')
```
